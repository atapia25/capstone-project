```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv ('ambient_15min.csv')
df2 = pd.read_csv ('system_15min.csv')
df3 = pd.read_csv ('atmos_15min.csv')
df4 = pd.merge(df2, df3, on='date', how='inner')
df = pd.merge(df1, df4, on='date', how='inner').set_index('date')
```

```python
df.head()
```

```python
t = range(0,15*len(df['temp_ambient']),15)
plt.plot(t, df['temp_ambient'],':', t, df['temp_system'], '--',t, df['temp_atmos_60'], t,df['temp_atmos_10'])
```

```python
a = np.array(df['temp_system'])
b = np.array(df['temp_ambient'])

ratio = a/(b+(b==0))

plt.ylim(-200,200)
plt.plot(t,ratio)
print(f"mean {np.mean(ratio)}")
```

```python
a = np.array(df['temp_system'])
b = np.array(df['temp_ambient'])

diff = a-b

plt.ylim(10,30)
plt.plot(t,diff)
print(f"mean {np.mean(diff)}, std {np.std(diff)}")
```

```python
plt.plot(t,df['temp_system'].rolling(window=100).mean().values, t, df['temp_system'])
```

```python
a = np.array(df['temp_system'])
b = np.array(df['temp_ambient'])

diff = a-b

plt.ylim(0,40)
plt.plot(t,a, t, b+np.mean(diff))

```

```python
c = np.array(b+np.mean(diff))

ratio = a/c
#system/ambient + mean of difference between system and ambient

plt.ylim(0,2)
plt.plot(t,ratio)
print(f"mean {np.mean(ratio)}")
```
